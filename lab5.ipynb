```python
df['Target'] = df['Open'].shift(-1)
df.dropna(inplace=True)
df
```

```python
train_df = df[:'2023-03-31']
test_df = df['2023-04-01':]
```

```python
import numpy as np
from torch import nn, Tensor
from torch.utils.data import TensorDataset, DataLoader

batch_size = 128
features_size = len(train_df.drop(['Target'], axis=1).columns)

train_data = TensorDataset(Tensor(np.array(train_df.drop(['Target'], axis=1))), Tensor(np.array(train_df[['Target']])))
test_data = TensorDataset(Tensor(np.array(test_df.drop(['Target'], axis=1))), Tensor(np.array(test_df[['Target']])))

train_loader = DataLoader(train_data, batch_size = batch_size)
test_loader = DataLoader(test_data, batch_size = batch_size)
```

```python
print(batch_size, features_size)
```

```python
hiddenSize = 256

class extract_tensor(nn.Module):
    def forward(self,x):
        # Output shape (batch, features, hidden)
        tensor, _ = x
        # Reshape shape (batch, hidden)
        return tensor[:, -1, :]

model = nn.Sequential(
    nn.RNN(19, hiddenSize, num_layers=1, batch_first=True),
    extract_tensor(),
    nn.Sequential(
        nn.Linear(hiddenSize, 128),
        nn.Linear(hiddenSize, 16),
        nn.Linear(hiddenSize, 1)
    )
)
```

```python
from torch import optim

loss_function = nn.MSELoss()

# Only SGD, SpareAdam and Adagrad support spare gradients
optimizer = optim.SGD(model.parameters(), lr=1e-3, momentum=0.9)
```

```python
from tqdm import tqdm

NUM_EPOCHS = 1

model.train() # put model in training mode

for epoch in range(NUM_EPOCHS):
  loop = tqdm(train_loader, position=0, leave=True)
  
  for (batch, labels) in loop:
    shape = batch.shape
    # batch = batch.reshape(shape[0], 1, shape[1])

    for input in batch:
      optimizer.zero_grad()
      output = model.forward(input)

      loss = loss_function(output, labels)
      loss.backward()

      optimizer.step()
```
